# Deteksi Marka Jalan

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Membaca gambar

```python
image = cv2.imread('Marka_Jalan.jpeg')
```

# Mengubah citra berwarna menjadi citra dalam skala keabuan

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

# Menerapkan efek Gaussian blur pada citra

```python
blur = cv2.GaussianBlur(gray, (5, 5), 0)
```

#  Melakukan deteksi tepi pada citra

```python
edges = cv2.Canny(blur, 190, 100)
```

#  Mendeteksi garis-garis menggunakan Transformasi Hough Probabilistik pada citra tepi

```python
lines = cv2.HoughLinesP(edges, 1, np.pi/150, threshold=60, minLineLength=20, maxLineGap=50)
```

# Menggambar garis-garis yang terdeteksi pada gambar asli

```python
if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(image, (x1, y1), (x2, y2), (0, 0, 255), 3)
```

# Menampilkan gambar

```python
cv2.imshow('Deteksi', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,5))
axs[0].imshow(image)
axs[1].plot(color='green')
plt.show
```
